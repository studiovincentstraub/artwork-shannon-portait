```python
import itertools
import numpy as np
import pandas as pd
```

```python
# declare size of matrix
arr_len = 30
arr_wid = 40

# create arr_eln x arr_wid matrix of zeros
arr = np.zeros((arr_len, arr_wid), dtype=int)

# fill matrix indicies based on defined binary pattern
vals = [['10'], [1, 39]], [['10'], [7, 33]]
# arr_row = 


d = []
row_idx = 0
for digits, counts in vals:
    # Build the pattern string for the row
    pattern = ''.join([str(digit) * count for digit, count in zip(digits, counts)])
    print("Pattern:", pattern)
    # Fill the row in the array with the pattern (truncate or pad to arr_wid)
    for col_idx in range(arr_wid):
        arr[row_idx, col_idx] = int(pattern[col_idx]) if col_idx < len(pattern) else 0
    print(arr[row_idx])
    row_idx += 1

df = pd.DataFrame(arr)

print(df.to_string(header=False, index=False))
```
